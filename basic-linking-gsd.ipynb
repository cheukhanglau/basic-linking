```python
import gsd.hoomd
import numpy as np
import pyknotid
import pyknotid.spacecurves
from pyknotid.spacecurves import Link as link
from pyknotid.invariants import alexander
np.random.seed(42)

# Function to read GSD file and extract data
def read_gsd_file(file_path):
    trajectory = gsd.hoomd.open(file_path, 'r')
    frames = [frame for frame in trajectory]  # Extract all frames
    return frames

# Path to your GSD file
file_path = 'chainlength5.gsd'

# Read the GSD file
frames = read_gsd_file(file_path)

# Process the first frame (or choose a specific frame)
frame = frames[0]

# Assuming the frame contains particles' positions in 3D space
points = frame.particles.position

# Since your original code expects data in a specific shape, we'll need to reshape the points
# Assuming mol_size is the same as the number of particles per molecule in your GSD file
mol_size = 5  # Update this value based on your specific data
data = points.reshape([-1, mol_size, 3])

print(data, data.shape)

stride = 1

data = data[:, ::stride, :]

print(data.shape)

# Assuming the simulation box side length is known
box_size = 53.7321  # This magic number is the simulation box side length
links = link.from_periodic_lines(data, shape=[box_size] * 3)
links.smooth(periodic=False)
a = links.linking_number(include_closures=True)

links.plot()

print('LINKING NUMBER: ', a)
```
